```
# Try different representations
def candidatePosition (originDataFrame):
    '''
    function invert the list of preference (in rank form) into nubers of 
    occurance of every element in rank at every position of rank 
       
    input:  originDataFrame - pandas.DataFrame with list as a feature
    output: candidatePositionDataFrame - pandas.DataFrame with columns(features)
            as a number occurance of the element at each position
    '''
    #[rowNum, colNum] = originDataFrame.iloc[0:,:-6].shape
    [rowNum, colNum] = originDataFrame.shape
    print('rowNum, colNum',rowNum, colNum)
    # create new dataframe fullfilled with zeros
    rankLength = len(eval(originDataFrame.loc[0][0]))
    data = np.zeros((rowNum,rankLength*rankLength), dtype =int)
    df = pd.DataFrame(data= data, columns=['{0}_pos{1}'.format(el,position) for el in range(1, rankLength+1) for position in range(1, rankLength+1)])
   
    for row in range(rowNum):
        print('row',row)
        for col in range(colNum):
            rank = originDataFrame.loc[row][col]
            rank1 = eval(rank) #str->int
            for el  in rank1:
                position = rank1.index(el)                   # find the index
                colName = str(el)+'_pos'+str(position+1)    # name the column in df
                df[colName][row] = df[colName][row] + 1   #  update the occurance number
                    
    candidatePositionDataFrame = df
    #candidatePositionDataFrame = df.join(ah.iloc[0:,-7:])#merge the borda, kemmeny and dodgson winner columns
    return candidatePositionDataFrame
```
