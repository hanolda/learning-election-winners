```
# Functions
def createScoreForPairsComparison(matrix, numCandidates):
    numCombination, pairToIndex = createPairsCombination(numCandidates)
    points = []    
    for i in range(len(matrix)):
        temp = []
        for j in range(numCombination):
            temp.append(0)
        points.append(temp)
     
    liste = len(matrix)
    for rowIndex in range(len(matrix)):
        row = matrix[rowIndex]
        for firstElementComp in range(len(row)-1):
            for secondElementCopm in range(firstElementComp+1, len(row)):
    #            print(row[firstElementComp],row[secondElementCopm])
                if type(row[secondElementCopm]) is list:
                    for listElements in row[secondElementCopm]:
                        if row[firstElementComp] > listElements:
                            points[rowIndex][pairToIndex["{}{}".format(listElements,row[firstElementComp])]] = 0#-1
                        else:
                            points[rowIndex][pairToIndex["{}{}".format(row[firstElementComp],listElements)]] = 1
                    for i in range(len(row[secondElementCopm])-1):
                        for j in range(i+1,len(row[secondElementCopm])):
                            if row[secondElementCopm][i]> row[secondElementCopm][j]:
                                points[rowIndex][pairToIndex["{}{}".format(row[secondElementCopm][j],row[secondElementCopm][i])]] = 0
                            else:
                                points[rowIndex][pairToIndex["{}{}".format(row[secondElementCopm][i],row[secondElementCopm][j])]] = 0
                                
                else:
                    if row[firstElementComp] > row[secondElementCopm]:
                        points[rowIndex][pairToIndex["{}_{}".format(row[secondElementCopm],row[firstElementComp])]] = 0# -1
                        points[rowIndex][pairToIndex["{}_{}".format(row[firstElementComp],row[secondElementCopm])]] = 1
                    else:
                        points[rowIndex][pairToIndex["{}_{}".format(row[firstElementComp],row[secondElementCopm])]] = 1

    return matrix, points
 
def createPairsCombination(numCandidates):
    # Get all combinations of [1, 2, 3]
    # and length 2
    comb = list(combinations(range(1,numCandidates+1), 2))
    comb_extension =[tuple(reversed(el)) for el in comb]
    comb.extend(comb_extension)
    
    
    numCombination = len(comb)
    # Print the obtained combinations
    pairToIndex = dict()
    counter = 0
    for i in comb:
        pairToIndex["{}_{}".format(i[0],i[1])]= counter
        #print("key", "{}{}".format(i[0],i[1]), 'val', counter)
        
        counter+=1
    return numCombination, pairToIndex
```
